```python
from datasets import DatasetDict
from data_processing.qmsum import load_qmsum_train_dataset, load_qmsum_test_dataset, prepare_qmsum_train, prepare_qmsum_test_ppl

"""
With DatasetDict, we can push multiple splits of the dataset to the Hugging Face Hub.
"""

# Load QMSum dataset
qmsum_train = load_qmsum_train_dataset(path="/home/yingqi/repo/QMSum/data/train.jsonl")
qmsum_train = prepare_qmsum_train(qmsum_train)
qmsum_test = load_qmsum_test_dataset(split='test')  # Test dataset is from long bench
qmsum_test = prepare_qmsum_test_ppl(qmsum_test)

# Create a DatasetDict
qmsum_dataset = DatasetDict({
    'train': qmsum_train,
    'test': qmsum_test
})

# Push to Hugging Face Hub
# qmsum_dataset.push_to_hub("ioeddk/qmsum")
# print("QMSum dataset pushed to Hugging Face Hub successfully!")
```

```python
IDX = 5
THREASHOLD = 0.001

print(qmsum_test[IDX]['answer'])
print()

# Print the answer for all train data
for i, example in enumerate(qmsum_train):
    if example['answer'].lower() in qmsum_test[IDX]['answer'].lower()[:int(len(qmsum_test[IDX]['answer']) * THREASHOLD)]:
        print(f"Example {i} Answer:", example['answer'])
```
